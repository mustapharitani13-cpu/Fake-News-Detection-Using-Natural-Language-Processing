```python
import pandas as pd

# Read CSV safely with the Python engine
df = pd.read_csv("/content/Fake_Real_Data.csv", engine='python', on_bad_lines='warn')

# Check first rows
print(df.head())
print("Total rows:", len(df))
```

```python
df['label_num']=df['label'].map({'Fake': 0, 'Real':1})
```

```python
# now convert Text column into vectors
import spacy
nlp=spacy.load('en_core_web_lg')
```

```python
doc=nlp("Top Trump Surrogate BRUTALLY Stabs Him In The...")
doc.vector
```

```python
# i want to do vectors to all rows so,i could do it by adding a column that stores vectors from the text and apply a sample function using lambda
df['text_vector']=df['Text'].apply(lambda x: nlp(x).vector)  # (x): every row in text / nlp(x): convert it to nlp then apply vector
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(
    df.text_vector.values,
    df.label_num,
    test_size=0.2,
    random_state=2025
)
```

```python
# when we give this to our classifier for the training it expecting 2D numpy array , So we need to convert this array from ths format to 2D numpy array
```

```python
# you can do it by using FUNCTION called     """            stack            """
# do it for x_train x_test
```

```python
# let's do it here
import numpy as np
x_train_2d=np.stack(x_train)
x_test_2d=np.stack(x_test)
```

```python
x_train_2d
```

```python
#ValueError: Negative values in data passed to MultinomialNB (input X).
# negative values are not acceptable in MultinomialNB
```

```python
# the solution can help you is to do scaling min-max
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_train_embad=scaler.fit_transform(x_train_2d)
scaled_test_embad=scaler.transform(x_test_2d) # we do it also for test but only transform NOT with fit.

from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(scaled_train_embad, y_train)
```

```python
y_pred=clf.predict(scaled_test_embad)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# knn algorithm
from sklearn.neighbors import KNeighborsClassifier
knn_clf=KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_clf.fit(scaled_train_embad, y_train)
```

```python
y_pred=knn_clf.predict(scaled_test_embad)
print(classification_report(y_test, y_pred))
```
